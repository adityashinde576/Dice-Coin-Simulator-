```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# Simulation 1: Dice Roll Experiment
# ---------------------------------------------------------
# Goal:
#   Simulate rolling a 6-sided dice 1000 times and visualize
#   the frequency distribution of outcomes (1 to 6).
# ---------------------------------------------------------

# Generate 1000 random dice rolls (values from 1 to 6)
dice = np.random.randint(1, 7, 1000)

# Plot histogram of dice outcomes
plt.hist(dice, bins=6, rwidth=0.8)
plt.title("Dice Roll Histogram")
plt.xlabel("Dice Value (1 to 6)")
plt.ylabel("Frequency")
plt.grid(True, axis='y')   # Horizontal grid lines only
plt.show()


# ---------------------------------------------------------
# Simulation 2: Coin Toss Experiment
# ---------------------------------------------------------
# Goal:
#   Simulate tossing a fair coin 1000 times and visualize
#   the frequency of Heads (1) and Tails (0).
# ---------------------------------------------------------

# Generate 1000 random coin tosses (0 = Tails, 1 = Heads)
coins = np.random.choice([0, 1], size=1000)

# Plot histogram of coin outcomes
plt.hist(coins, bins=2, rwidth=0.5)
plt.title("Coin Toss Histogram")
plt.xlabel("Outcome (0 = Tails, 1 = Heads)")
plt.ylabel("Frequency")
plt.grid(True, axis='y')   # Horizontal grid lines only
plt.show()
```
